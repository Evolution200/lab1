```python
import torch
import numpy as np
import matplotlib.pyplot as plt
#import cv2
import gc
from PIL import Image
from transformers import pipeline
```

```python
# open and show raw images
# you can examine the directory for the images

img_path = '/share/lab5/data/test3.jpg'
raw_image = Image.open(img_path)
raw_image.show()
```

```python
# Create a mask-generation pipeline from the model /share/lab5/sam/, note the ending "/" in the path name

generator = pipeline("mask-generation", model="/share/lab5/sam/", device=0)
```

```python
# Using the pipeline to generate the mask

outputs = generator(raw_image, points_per_batch=64)
masks = outputs["masks"]
```

```python
# Examine the mask data structure
print(type(masks))
print(masks[0].shape)
masks[0]
```

```python
## some utility functions to handle the masks

def show_mask(mask, ax, random_color=True):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30 / 255, 144 / 255, 255 / 255, 0.6])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)
    del mask
    gc.collect()

def show_masks_on_image(raw_image, masks):
  plt.imshow(np.array(raw_image))
  ax = plt.gca()
  ax.set_autoscale_on(False)
  for mask in masks:
    show_mask(mask, ax=ax, random_color=True)
  plt.axis("off")
  plt.show()
  del mask
  gc.collect()
```

```python
show_masks_on_image(raw_image, masks)
```

```python
# Another example 

img_path = '/share/lab5/data/test8.jpg'
raw_image = Image.open(img_path)
raw_image.show()

outputs = generator(raw_image, points_per_batch=64)
masks = outputs["masks"]
show_masks_on_image(raw_image, masks)
```

```python
#### Your Task ####
# show only the mask with the largest area
# hint: you need to examine the masks data structure to see how to calcluate the area.
# if there are tow masks with the same area size, show either one
import numpy as np

def find_largest_mask(masks):
    # Initialize maximum area and index
    max_area = 0
    max_index = 0

    # Iterate over all masks
    for i, mask in enumerate(masks):
        # Calculate area of the current mask
        area = np.sum(mask)

        # If this area is larger than the current maximum, update maximum
        if area > max_area:
            max_area = area
            max_index = i

    # Return the mask with the largest area
    return masks[max_index]

def show_largest_mask_on_image(raw_image, masks):
    # Find the largest mask
    largest_mask = find_largest_mask(masks)

    plt.imshow(np.array(raw_image))
    ax = plt.gca()
    ax.set_autoscale_on(False)
    show_mask(largest_mask, ax=ax, random_color=True)
    plt.axis("off")
    plt.show()
    del largest_mask
    gc.collect()

show_largest_mask_on_image(raw_image, masks)
```
